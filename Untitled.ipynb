```python
from ultralytics import YOLO
import os

# === CONFIGURATION ===
MODEL_PATH = '/teamspace/studios/this_studio/runs/pose/train19/weights/best.pt'
TEST_IMAGES_DIR = '/teamspace/studios/this_studio/dataset_thermal_split/dataset_thermal_split/images/test/'
SAVE_DIR = '/teamspace/studios/this_studio/predicted_samples_per_class'

# === LOAD MODEL ===
model = YOLO(MODEL_PATH)

# === SETUP ===
os.makedirs(SAVE_DIR, exist_ok=True)
saved_classes = set()
class_names = model.names

# === RUN PREDICTION ON TEST IMAGES ===
results = model.predict(source=TEST_IMAGES_DIR, stream=True, conf=0.5, save=False)

for r in results:
    if r.boxes is None or len(r.boxes.cls) == 0:
        continue

    class_ids = r.boxes.cls.cpu().numpy().astype(int)

    for class_id in class_ids:
        class_name = class_names[class_id]

        if class_name not in saved_classes:
            save_path = os.path.join(SAVE_DIR, f"{class_name}.jpg")
            r.save(filename=save_path)  # Save annotated image
            saved_classes.add(class_name)
            print(f"✔️ Saved: {save_path}")
            break  # Move to next image after saving one class

    if len(saved_classes) == len(class_names):
        break  # Exit once all classes are covered

print(f"\n✅ Done. Saved one prediction image per class in: {SAVE_DIR}")
```

```python
from ultralytics import YOLO
import os
import shutil
import pandas as pd

# === CONFIGURATION ===
MODEL_PATH = '/teamspace/studios/this_studio/runs/pose/train19/weights/best.pt'
TEST_IMAGES_DIR = '/teamspace/studios/this_studio/dataset_thermal_split/dataset_thermal_split/images/test/'
SAVE_DIR = '/teamspace/studios/this_studio/yoloresult/predicted_samples_per_class_best'
HTML_OUTPUT = '/teamspace/studios/this_studio/yoloresult/prediction_table_best.html'
CSV_OUTPUT = '/teamspace/studios/this_studio/yoloresult/prediction_table_best.csv'

# === LOAD MODEL ===
model = YOLO(MODEL_PATH)
os.makedirs(SAVE_DIR, exist_ok=True)
class_names = model.names

# === Track best prediction per class ===
best_predictions = {}

# === PREDICT ON TEST IMAGES ===
results = model.predict(source=TEST_IMAGES_DIR, stream=True, conf=0.9, save=False)

for r in results:
    if r.boxes is None or len(r.boxes.cls) == 0:
        continue

    image_path = r.path
    class_ids = r.boxes.cls.cpu().numpy().astype(int)
    confidences = r.boxes.conf.cpu().numpy()

    for class_id, conf in zip(class_ids, confidences):
        class_name = class_names[class_id]

        # Only update if this is the highest confidence for the class
        if class_name not in best_predictions or conf > best_predictions[class_name]['conf']:
            best_predictions[class_name] = {
                'class_name': class_name,
                'conf': float(conf),
                'image_path': image_path,
                'result_obj': r
            }

# === SAVE BEST PREDICTIONS ===
html_rows = []
csv_rows = []

for class_name, entry in best_predictions.items():
    base_name = class_name.replace(" ", "_")

    original_img_path = os.path.join(SAVE_DIR, f"{base_name}_original.jpg")
    predicted_img_path = os.path.join(SAVE_DIR, f"{base_name}_predicted.jpg")

    # Save images
    shutil.copy(entry['image_path'], original_img_path)
    entry['result_obj'].save(filename=predicted_img_path)

    # Add to HTML + CSV
    html_rows.append(
        f"<tr>"
        f"<td><img src='{original_img_path}' height='200'></td>"
        f"<td><img src='{predicted_img_path}' height='200'></td>"
        f"<td>{class_name}</td>"
        f"<td>{round(entry['conf'], 4)}</td>"
        f"</tr>"
    )

    csv_rows.append({
        'Class': class_name,
        'Confidence': round(entry['conf'], 4),
        'Original_Image': original_img_path,
        'Predicted_Image': predicted_img_path
    })

# === SAVE HTML ===
html_content = f"""
<html>
<head>
    <title>Best YOLOv8 Pose Predictions</title>
</head>
<body>
    <h2>Best Prediction per Class</h2>
    <table border="1" cellpadding="10">
        <tr>
            <th>Original Image</th>
            <th>Predicted Image</th>
            <th>Class</th>
            <th>Confidence</th>
        </tr>
        {''.join(html_rows)}
    </table>
</body>
</html>
"""

with open(HTML_OUTPUT, 'w') as f:
    f.write(html_content)

# === SAVE CSV ===
df = pd.DataFrame(csv_rows)
df.to_csv(CSV_OUTPUT, index=False)

print(f"\n✅ Best predictions saved")
print(f"HTML: {HTML_OUTPUT}")
print(f"CSV : {CSV_OUTPUT}")
```
